**hierarchical clustering-1**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Create a sample dataset with 2D data
data = {
    'Feature 1': [1, 2, 2, 3, 6, 7, 8, 8, 25, 24],
    'Feature 2': [1, 1, 2, 2, 6, 7, 8, 8, 25, 25]
}
df = pd.DataFrame(data)

# Perform hierarchical clustering
clustering = AgglomerativeClustering(n_clusters=3)  # Change the number of clusters as needed
clusters = clustering.fit_predict(df)

# Create a scatter plot to visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature 1'], df['Feature 2'], c=clusters, cmap='viridis', marker='o', edgecolor='k')
plt.title('Hierarchical Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

**hierarchical clustering-2**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Create a larger sample dataset with 2D data
np.random.seed(42)
cluster_1 = np.random.normal(loc=2, scale=0.5, size=(50, 2))  # 50 points around (2, 2)
cluster_2 = np.random.normal(loc=7, scale=0.5, size=(50, 2))  # 50 points around (7, 7)
cluster_3 = np.random.normal(loc=12, scale=0.5, size=(50, 2)) # 50 points around (12, 12)
cluster_4 = np.random.normal(loc=20, scale=0.5, size=(50, 2)) # 50 points around (20, 20)

# Combine the clusters into a single DataFrame
df = pd.DataFrame(np.vstack([cluster_1, cluster_2, cluster_3, cluster_4]), columns=['Feature 1', 'Feature 2'])

# Perform hierarchical clustering
clustering = AgglomerativeClustering(n_clusters=4)  # Change the number of clusters as needed
clusters = clustering.fit_predict(df)

# Create a scatter plot to visualize the clusters
plt.figure(figsize=(10, 8))
plt.scatter(df['Feature 1'], df['Feature 2'], c=clusters, cmap='viridis', marker='o', edgecolor='k')
plt.title('Hierarchical Clustering Results with More Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

**K-means clustering**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Perform K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(df)

# Create a scatter plot to visualize the clusters
plt.figure(figsize=(10, 8))
plt.scatter(df['Feature 1'], df['Feature 2'], c=clusters, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering Results with More Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
#plt.colorbar(label='Cluster')
plt.legend()
plt.grid()
plt.show()
```
